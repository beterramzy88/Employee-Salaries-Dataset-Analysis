## Q2. Descriptive Statistics
Calculate basic statistics `mean`, `median`, `mode`, `minimum`, and `maximum` salary, determine the `range of salaries`, and find the `standard deviation`.

```python
# Calculate mean, median, mode, minimum, and maximum salary
mean_salary = df['TotalPay'].mean()
median_salary = df['TotalPay'].median()
mode_salary = df['TotalPay'].mode().iloc[0]  # Selecting the first mode in case there are multiple modes
min_salary = df['TotalPay'].min()
max_salary = df['TotalPay'].max()
std_dev_salary = df['TotalPay'].std()

# Determine the range of salaries
salary_range = max_salary - min_salary

# Print the results
print("Mean Salary:", mean_salary)
print("Median Salary:", median_salary)
print("Mode Salary:", mode_salary)
print("Minimum Salary:", min_salary)
print("Maximum Salary:", max_salary)
print("Range of Salaries:", salary_range)
print("Standard Deviation of Salary:", std_dev_salary)
```

## Q3 Data Cleaning
 Handle missing data by suitable method with explain why you use it.

### 1\. Numeric Columns ('BasePay', 'OvertimePay', 'OtherPay', 'Benefits'):

*   **Handling Strategy:** Imputation with Mean
*   **Explanation:** Imputing missing values with the mean is a common practice for numeric columns. It helps maintain the overall distribution of the data and is a simple approach when the missing values are expected to be missing at random. Using the mean ensures that the imputed values are representative of the typical values in each respective column.

```python
df['BasePay'].fillna(df['BasePay'].mean(), inplace=True)
df['OvertimePay'].fillna(df['OvertimePay'].mean(), inplace=True)
df['OtherPay'].fillna(df['OtherPay'].mean(), inplace=True)
df['Benefits'].fillna(df['Benefits'].mean(), inplace=True)
```

### 2\. Categorical Columns ('EmployeeName' and 'JobTitle'):

*   **Handling Strategy:** Imputation with Mode
*   **Explanation:** Filling missing values in categorical columns with the mode (most frequent value) is a standard approach. It is suitable when the missing values can be reasonably replaced by the most common values in the respective columns. This helps preserve the categorical nature of the data and ensures that imputed values are consistent with the existing distribution.


```python
df['EmployeeName'].fillna(df['EmployeeName'].mode().iloc[0], inplace=True)
df['JobTitle'].fillna(df['JobTitle'].mode().iloc[0], inplace=True)
```

### 3\. Dropping Columns with Many Missing Values ('Notes', 'Status'):

*   **Handling Strategy:** Column Dropping
*   **Explanation:** Since these columns have a substantial number of missing values, and they may not contribute significantly to the analysis, dropping them is a reasonable choice. This simplifies the dataset and avoids imputing a large number of missing values.

```python
df.drop(['Notes', 'Status'], axis=1, inplace=True)
```

## Q4. Basic Data Visualization
Create histograms or bar charts to visualize the distribution of salaries, and use pie charts to represent the proportion of employees in different departments.

### 4.1 Create histograms or bar charts to visualize the distribution of salaries

```python
# Set the style for seaborn
sns.set(style="whitegrid")

# Create a histogram for the distribution of salaries
plt.figure(figsize=(12, 6))
sns.histplot(df['TotalPay'], bins=50, kde=True, color='skyblue')
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay')
plt.ylabel('Frequency')
plt.show()
```

### 4.2 use pie charts to represent the proportion of employees in different departments.

```python
# Create a pie chart for the proportion of employees in different departments
plt.figure(figsize=(10, 8))
department_counts = df['JobTitle'].value_counts()
top_departments = department_counts.head(15)  # Displaying top 15 departments for better visualization
plt.pie(top_departments, labels=top_departments.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

## Q5. Grouped Analysis
Group the data by one or more columns and calculate summary statistics for each group, and compare the average salaries across different groups.

### 5.1 Group the data by one or more columns
I group by the year column because it gives me minimum number of groups than grouping by other columns.

```python
# Group the data by 'Year'
grouped_data = df.groupby('Year')

# Get the number of groups
num_groups = len(grouped_data)

# Get the names of the columns
columns_names = grouped_data.head(0).columns

# Display the number of groups and column names
print(f"Number of Groups: {num_groups}")
print(f"Column Names: {columns_names}")
```

### 5.2 calculate summary statistics for each group

```python
# Calculate summary statistics for every column within each group
summary_statistics_by_year = grouped_data.describe()

# Display the summary statistics for every column in a more organized format
for column in summary_statistics_by_year.columns.levels[0]:
    print(f"\nSummary Statistics for {column} by Year:")
    print(summary_statistics_by_year[column].to_string())
    print("\n" + "="*80)  # Add a separator for better visibility
```

### 5.3 compare the average salaries across different groups.


```python
# Group the data by 'Year' and calculate average salary for each group
grouped_by_year = df.groupby('Year')[['TotalPay']].mean()

# Iterate over each group (year) and compare average salaries
for year, data in grouped_by_year.iterrows():
    print(f"\nYear: {year}")
    print("Average TotalPay:", data['TotalPay'])
```
